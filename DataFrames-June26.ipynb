# Introduction to DataFrames

- A dataframe represents a tabular data (table)
- Contains rows and columns
- By default, rows are referred by index (0 to n-1)
- And columns are also referred by their index
- Rows are at axis=0 (x-axis), and columns are at axis=1 (y-axis)

````python
# imports
import numpy as np
import pandas as pd
````

````python
data = {
    'city':['hyd', 'del', 'mum', 'che', 'kol'],
    'state':['telangana', 'delhi', 'maharastra', 'tamilnadu', 'west bengal'],
    'population':[15000000, 18000000, 14000000, 16000000, 12000000]
}
````

````python
pd.DataFrame(data)
````

````python
customer_visits = np.random.randint(10, 25, (3, 5))
customer_visits
````

````python
pd.DataFrame(customer_visits)
````

````python
pd.DataFrame(customer_visits, index=['Morning', 'Afternoon', 'Evening'])
````

````python
pd.DataFrame(customer_visits, 
             index=['Morning', 'Afternoon', 'Evening'],
            columns=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
````
