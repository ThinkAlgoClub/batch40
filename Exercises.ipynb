## Counting attempts of a password

```python
correct_password = "python123"
attempts = 0

while attempts < 3:
    password = input("Enter a password: ")
    if password == "":
        print("No password entered")
        continue

    if password == correct_password:
        print("Password is correct")
        break
    else:
        print("Password entered is wrong")
        attempts += 1

if attempts == 3:
    print("Account is locked")
else:
    print("Account logged in")
```

## Finding the perfect number

```python
total = 0
n = int(input("Enter an integer: "))

for i in range(1, n):
    if n % i == 0:
        total += i

if n == total:
    print(n, "is a perfect number")
else:
    print(n, "not a perfect number")
```

## Finding strong number

```python
import math

total = 0
n = int(input("Enter an integer: "))
temp = n

while temp > 0:
    digit = temp % 10   # last digit (remainder)
    total += math.factorial(digit)
    temp = temp // 10

if n == total:
    print("Strong number")
else:
    print("Not a strong number")
```

## Find the prime factors

```python
n = int(input("Enter an integer: "))

# divide by two (2)
while n % 2 == 0:
    print(2, end=' ')
    n = n // 2

# start with 3 and then remaining odd numbers
i = 3
while i * i <= n:
    while n % i == 0:
        print(i, end=' ')
        n = n // i
    i += 2
print(n)
```
