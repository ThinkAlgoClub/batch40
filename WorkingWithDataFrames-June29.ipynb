## FILTERING A DATAFRAME

````python
employees = pd.read_csv("employees.csv", parse_dates=["Start Date"])
employees.head(3)
````

````python
employees.info()
````

#### `astype()` method to convert data types

````python
employees["Mgmt"] = employees["Mgmt"].astype(bool)
employees.info()
````

````python
# employees["Salary"] = employees["Salary"].astype(int) # IntCastingNaNError - integer type
````

````python
employees.head(3)
````

````python
# use fillna(value) method to to replace NaN values to specified value
````

````python
employees["Salary"].fillna(0)
````

employees.Salary

````python
employees["Salary"] = employees["Salary"].fillna(0).astype(int)
````

````python
employees.head(3)
````

````python
employees.info()
````

````python
len(employees)
````

````python
employees.nunique()
````

````python
employees["Gender"] = employees["Gender"].astype("category")
employees.info()
````

````python
employees["Team"] = employees["Team"].astype("category")
employees.info()
````

#### Conditions to apply for filtering

````python
employees.head(10)
````

````python
# a condition to create a boolean Series
employees["First Name"] == "Maria"
````

````python
# which can be used to filter the rows
marias = employees["First Name"] == "Maria"
# filter the dataframe with the condition
employees[marias]
````

````python
"Finance" != "Engineering"
````

````python
employees[employees["Team"] == "Finance"]
````

````python
finance = employees["Team"] == "Finance"
employees[finance]
````

````python
# employees[employees["Salary"] > 100000] # OR
high_salaries = employees["Salary"] > 100000
employees[high_salaries]
````

````python
is_female = employees["Gender"] == "Female"
is_biz_dev = employees["Team"] == "Business Dev"
# combine the two filter using & (and) - both the conditions must be met
employees[is_female & is_biz_dev]
````

````python
# Mgmt column is a boolean column, we can directly use it to filter the data
is_manager = employees["Mgmt"]
employees[is_manager]
````

````python
# Inversion operator - ~
# Inverts True to False, and False to True
````

````python
employees[~is_manager]
````

````python
employees[~(employees["Salary"] > 100000)]
````

````python
# The or operator is |
sales = employees["Team"] == "Sales"
marketing = employees["Team"] == "Marketing"
legal = employees["Team"] == "Legal"
employees[sales | marketing | legal]
````

````python
# isin() method instead of '|' opeators
star_teams = ["Sales", "Marketing", "Legal"]
all_star_teams = employees["Team"].isin(star_teams)
employees[all_star_teams]
````

````python
higher_than_80k = employees["Salary"] >= 80000
lower_than_100k = employees["Salary"] <= 100000
employees[higher_than_80k & lower_than_100k]
````

````python
# use between() method for numeric and date columns
salary_80_100k = employees["Salary"].between(80000, 100000)
employees[salary_80_100k]
````

````python
# use the between() with dates
folkes_80k = employees["Start Date"].between(
    left="1980-01-01",
    right="1990-01-01"
)
employees[folkes_80k]
````
