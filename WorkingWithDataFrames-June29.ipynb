```python
days_of_war = pd.read_csv(
    "revolutionary_war.csv",
    parse_dates=["Start Date"],    # converting the column to datetime
    usecols=["Start Date"],        # choose the columns to be in a dataframe
).squeeze()                        # if the dataframe has one column it can be squeezed to a series
days_of_war
```

```python
days_of_war.info()
```

```python
days_of_war = days_of_war.dropna() # dropna() to drop the missing values
```

```python
from datetime import datetime
def day_of_week(date):
    return date.strftime("%A")

# days_of_war.dropna().apply(day_of_week)
days_of_war.apply(day_of_week) # use the apply() method to apply a UDF on a Series
```

```python
days_of_war = days_of_war.dropna().apply(day_of_week)
days_of_war
```

```python
days_of_war.value_counts()   # count the occurances of each value in the SEries
```
