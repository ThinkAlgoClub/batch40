# The `if` construct/statement

- Conditional branching construct
- Every branch is a block of code to be executed
- However, the if statement executes code only from ONE branch or NO branch
- The if statement evaluates the branches in sequence

Syntax:
```
if bool_exp:
    block
[elif bool_exp:
    block]...
[else:
    block]
```

- There are three clauses in a `if` statement: if, elif, else
- if clause: first clause, required, only one 'if' clause - 1
- elif clause: optional, but we can have more than one 'elif' clause - 0 to N
- else clause: optional, only one - 0 or 1
- if and elif clauses: have bool_exp to evaluate
- The 'if' statement evaluates the bool_expressions of the clauses in sequence
      - if clause bool_exp is True; then it executes statements only from if clause, then ends the if statement
      - if clause bool_epx is False; then it checks the bool_exp of first elif clause, and if it is True, it executes the statements from elif clause, then ends the if statement
      - if caluse bool_exp is False and first elif clause bool_exp is False, then it goes to 2nd elif clause and if it is True it executes statements from the 2nd elif clause then end the if statement
      - if all the clauses are False, then it exeuctes statements from 'else' clause

- The block is a set of `indented` statements
- All the statements should be indented equally
- We indent statements when those statements are part of a clase: if, elif, else, def, with, for, while, class,....

````python
# nested ifs - one inside another
EXP = 3
EDU = 'mba'

input_exp = int(input("Enter your experience: "))
input_edu = input("Enter your education: ")

if input_exp >= EXP:
    if input_edu == EDU:
        print("Can apply for the job.")
    else:
        print("Education is not sufficient")
else:
    if input_edu == EDU:
        print("Not sufficient experience")
    else:
        print("You don't meet educational qualification and experience")
    
````

````python
EXP = 3
EDU = 'mba'

input_exp = int(input("Enter your experience: "))
input_edu = input("Enter your education: ")

if input_exp >= EXP and input_edu == EDU:
    print("Can apply for the job.")
elif input_exp >= EXP and input_edu != EDU:
    print("Education is not sufficient")
elif input_exp < EXP and input_edu == EDU:
    print("Not sufficient experience")
else:
    print("You don't meet educational qualification and experience")
````
