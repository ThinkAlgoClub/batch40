## Parsing strings to datetime values using `strptime` function

- The `strptime` is a datetime constructor's method
- Using this we can parse datetime values from a string
- The `p` in the method name stands for `parsing`

#### **Syntax for strptime**
```
datetime.strptime(datetime_str, format_str)
```

#### `format_str` codes:
- %y --> two digit year number
- %Y -> four digit year number
- %m -> month number (1 - 12)
- %d -> day of month (1 - 31)
- %H -> hour number (0 - 23), 24-hour format
- %M -> minute number (0 - 59)
- %S -> second number (0 - 59)

````python
dt = "2026-06-14"
dt = datetime.strptime(dt, "%Y-%m-%d")
print(dt)
````

````python
dt2 = "15 of 6th month, 2029"
dt2 = datetime.strptime(dt2, "%d of %mth month, %Y")
print(dt2)
````

````python
dt3 = "On the 18th day of 12th month in the year 2005, at 11 o clock and 30 minutes we had a conversation."
````

````python
dt3= datetime.strptime(dt3, "On the %dth day of %mth month in the year %Y, at %H o clock and %M minutes we had a conversation.")
print(dt3)
````

### Using `strftime` to convert datetime values to string

- We can format the datetime values as string
- We use `strftime()` method, where `f` stands for `format
#### **Syntax**
```
datetime_object.strftime(format_str)
```

#### `format_str` codes
```
%y - two digit year number
%Y - four digit year number
%m - month number
%d - day number of the month
%H - hour number (24-hour format)
%I - hour number (12-hour format)
%M - minute number
%S - second number
%p - AM/PM Specifier
%a - short weekday name (Sun, Mon, Tue, ...)
%A - full weekday name (Sunday, Monday, Tuesday, ...)
%b - short month name (Jan, Feb, Mar, ...)
%B - full month name (January, February, March, ...)
```

````python
one_day = datetime(2013, 5, 27, 8, 35, 30)
print(one_day)
````

````python
one_day.strftime("(%A) %b %d, %Y")
````

````python
another_day = datetime(2012, 10, 15, 17, 35, 30)
print(another_day)
````

````python
another_day.strftime("%B %d, %Y - (%a)")
````

````python
another_day.strftime("%Y/%m/%d (%A) - %H:%M")
````

````python
another_day.strftime("%Y/%m/%d (%A) - %I:%M %p")
````

````python
another_day.strftime("Date: %B %d, %Y - Time: %I:%M:%S %p")
````

## The `timedelta` object - span of time

- The `datetime` module contains a `timedelta` constructor to create time spans
- These time spans are internally differences between two values

````python
# importing timedelta contructor
from datetime import timedelta
````

#### **Syntax**
```
timedelta([weeks,] [days,] [hours,] [minutes,] [seconds,] [milliseconds,] [microseconds])
```

````python
project_duration = timedelta(weeks=6)
start_date = date(2026, 6, 15)
end_date = start_date + project_duration
print(end_date)
````

````python
task_span = timedelta(hours=3, minutes=30)
start_time = datetime(2026, 6, 11, 12, 5)
end_time = start_time + task_span
print(end_time.strftime("End time: %I:%M:%S %p"))
````

- The `timedelta` returns values in days, seconds, and microseconds

````python
type(task_span)
````

##### difference between two date values or between to datetime values creates a timedelta object

````python
bd_date = date(2026, 1, 23)
today = date.today()
````

````python
Num_of_days = today - bd_date
print(Num_of_days)
````

````python
if today > bd_date:
    print("This year birthday is already celebrated!")
else:
    print("Birthday is coming in the future this year")
````

````python
now = datetime.now()
target_dt = datetime(2026, 10, 15, 9, 30)
print(target_dt - now)
````
