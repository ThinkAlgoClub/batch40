### Comparison operators

````python
arr = np.array([1, 2, 3, 4, 5])
arr >= 3
````

````python
arr == 3
````

````python
arr > 0
````

````python
## any() and all() functions
````

````python
np.any(arr == 3)   # if at least one value is True it returns True
````

````python
np.all(arr == 3)  # if any value is False then it returns False
````

````python
np.all(arr > 0)  # returns True when all the values are True
````

### Conditional Selection (where)

````python
arr = np.array([10, 20, 30, 40])
result = np.where(arr > 25, 1, 0)   # 1 represent True; 0 represents False
result
````

### Non-zero Selection

````python
arr = np.array([3, 0, 1, 4, 0])
indexes = np.nonzero(arr)
arr[indexes]
````

### Stacking arrays

````python
a = np.array([10, 20, 30])
b = np.array([15, 25, 35])
````

````python
# vertical stacking
np.vstack((a, b))
````

````python
# horizontal stacking
np.hstack((a, b))
````

````python
# column stacking
np.column_stack((a, b))
````

### Matrix multiplication

### Inverse Matrix

````python
A = np.array([
    [4, 5],
    [10, 7]
])

inv_A = np.linalg.inv(A)
inv_A
````

````python
A @ inv_A
````
