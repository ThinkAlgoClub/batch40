# LOOP Construct/Statements

- A loop construct executes a block (code) repeatedly based on condition or collection/count.

- Two types loops:
    - Collection based loops - `for`
    - Conditional loops - `while`

## The `for` loop construct
- Collection (count of collection) based loop
- The loop executes the block for each item in the collection.
- Each execution is called an iteration.
- And a collection never has infinite number of items.

##### Syntax:
```
for var in collection:
    block
```

- The loop iterates for each item in the collection
- In each iteration, it assigns the next item to the var, and then executes the block
