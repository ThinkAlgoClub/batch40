### Variables
- Location in memory that can store a value or an object
- We can assume this location as a container, a container to store a value or an object
- To store data temporarily
- Syntax:
    - var = value
    - var = expression
    - container = value
    - container = expression
    - var = object
    - container = object
- The value/expression/object on the right-side of the assingment operator (=) is stored in the variable/container that is on the left-side of the assignment operator(=), not vice versa.

#### Rules for naming variables
1. A variable name is case-sensitive
2. A variable name can contain letters (upper and lower case), numbers and underscore(_) character
3. Use descriptive names for variable
4. Don't begin a variable with a number
5. Don't use keywords for variable names

```python
# Case-sensitive
x = 20
X = 25
```

```python
print(x)
print(X)
```

```python
# use letters, numbers and underscore
sales_2025 = 8500000
roi = 0.12
term = 240
Tax_percent_2024 = 0.28
```

```python
sub1 = 100
sub2 = 98
_3sub = 99
# 4sub - This is an error
```

```python
# can't use keywords for variable names
fi = 100
```

#### Line-continuation character
- \ (backslash) is the line continuation character
- Wherever you want to break a line, code \ and hit Return (Enter) key

```python
1 + 10 + 2 + 20 + 3 + 30 + 4 + 40 + 5 + 50 + 6 + 60 + 7 + 70

1 + 10 \
  + 2 + 20 \
  + 3 + 30 \
  + 4 + 40 \
  + 5 + 50 \
  + 6 + 60 \
  + 7 + 70
```
