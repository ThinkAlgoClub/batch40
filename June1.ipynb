# Working with Lists

- A list is a collection of sequential items
- All the items in a list are sequential and they indexed by index numbers ( 0 to n-1 )

- **Syntax for creating a list**:
```
[item1, item2, item3, item4, ...]
```

- Each item can be of any datatype

````python
# creating a list
[42.3, 44.7, 40.9, 43.7, 45.9]
````

````python
# list of scores
[99, 100, 98, 85, 94, 97]
````

## Accessing items from a list
- Every item in a list has an index number, and index numbers start from 0
- First item index is 0, second item index is 1, third item index is 2, and so on and so forth
- Similar there are negative indexes to get items from the end of the list
- Last item index number is -1, second last item is -2, third last item is -3, so on and so forth
- Use index number to get an item or 

- *Syntax for accessing items from a list*:
```
list_name[index_number]
```

- Here, [] are imply index operator

### Everything is an object in Python
- Every object has two aspects:
    - Attributes (characteristics, features, property)
        - object.attribute (get its value)
        - object.attribute = new_value (set its value)
    - Methods (functions, actions)
        - object.method([args])
    - Methods affect the attributes
