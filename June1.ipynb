## Accessing items from a list
- Every item in a list has an index number, and index numbers start from 0
- First item index is 0, second item index is 1, third item index is 2, and so on and so forth
- Similar there are negative indexes to get items from the end of the list
- Last item index number is -1, second last item is -2, third last item is -3, so on and so forth
- Use index number to get an item or 

- *Syntax for accessing items from a list*:
```
list_name[index_number]
```

### sort([key=function][, reverse=False])
- Sort the items in a list in ascending order
- The optional 'key' arguments is used to specify a function before sorting

````python
cities = ['bhu', 'Mum', 'blr', 'bhu', 'tri', 'che', 'del', 'amr', 'hyd', 'kol', 'kol', 'kol']
cities.sort()
````

````python
print(cities)
````

````python
cities.sort(key=str.lower) # cities.sort(key=str.upper)
print(cities)
````

````python
cities.sort(key=str.upper, reverse=True) # cities.sort(key=str.lower)
print(cities)
````

#### sorted(list[, key=function]) - built-in function
- This function returns a new list without modifying the original list

````python
cities = ['bhu', 'Mum', 'blr', 'bhu', 'tri', 'che', 'del', 'amr', 'hyd', 'kol', 'kol', 'kol']
sorted_cities = sorted(cities, key=str.lower)
print("Old list:", cities)
print("New list:", sorted_cities)
````

### List slicing
- *Syntax*
```
list[start:end:step]
```

````python
numbers = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60]
````

````python
numbers[0:4]
````

````python
numbers[0::2] # from first value, alternate values
````

````python
numbers[1::2] # from second value, alternate values
````

````python
numbers[3:7] # from index 3 to index 6, 7 is exlusive
````

````python
numbers[::-1] # reverse the items in a list and returns it
````

````python
numbers[::3]
````
