# LOOP Construct/Statements

- A loop construct executes a block (code) repeatedly based on condition or collection/count.

- Two types loops:
    - Collection based loops - `for`
    - Conditional loops - `while`

## The `for` loop construct
- Collection (count of collection) based loop
- The loop executes the block for each item in the collection.
- Each execution is called an iteration.
- And a collection never has infinite number of items.

##### Syntax:
```
for <var> in <collection>:
    block
```

- The loop iterates for each item in the collection
- In each iteration, it assigns the next item to the var, and then executes the block

#### Examples

````python
for x in [10, 85, -90, 38]:
    print("Hello world!")
    print("------------")
````

````python
for i in ["Raju", 35, 120000.0]:
    print(i)
````

````python
for temp in (41.5, 43.8, 44.3, 45, 44.9):
    print("Temperature:", temp)
````

````python
for temp in (41.5, 43.8, 44.3, 45, 44.9):
    print(temp, end=", ")
````

````python
for l in "HELLO":
    print(l)
````

````python
for l in "HELLO":
    print(l, end=' ')
````

````python
total = 0
for i in [1, 2, 3, 4, 5]:
    total = total + i
print(total)
````

### The range() iterator function
- The range() function gives a series (sequence) of numbers, but not all at once
- This is an iterator function which gives one number from the series for each iteration in a loop

### Syntax:
`range([start,] stop [, step])`
- `start` (optional): 0 is default
- `step` (optional): 1 is default
- `stop` (required): based on `start` and `step` value it can either positive or negative

The iterator (range() function) gives numbers from `start` value till `stop` value, where start is inclusive and stop is `exclusive`, by continuously adding the `step` value to the `start` value

#### range(stop)
- start is 0 and step 1; stop is +ve

###### Illustrations
- range(5)        # 0, 1, 2, 3, 4
- range(8)        # 0, 1, 2, 3, 4, 5, 6, 7
- range(3)        # 0, 1, 2

````python
for i in range(5):
    print(i, end=' ')
````

````python
for i in range(8):
    print(i, end=' ')
````

````python
for i in range(3):
    print(i, end=' ')
````

````python
# Adding 0 to 10
total = 0
for i in range(11):
    total = total + i
print(total)
````

#### range(start, stop)
- step 1; start < stop

##### Illustrations
- range(5, 10)      # 5, 6 7, 8, 9
- range(-10, -5)    # -10, -9, -8, -7, -6
- range(-3, 3)      # -3, -2, -1, 0, 1, 2
- range(3, -3)      # WRONG: START > STOP

````python
for i in range(5, 10):
    print(i, end=' ')
````

````python
for i in range(-10, -5):
    print(i, end=' ')
````

````python
total = 0
for i in range(1, 101):
    total = total + i
print(total)
````

#### range(start, stop, step)
- step +ve; start < stop
- step -ve; start > stop

##### Illustrations
- range(5, 50, 10)     # 5, 15, 25, 35, 45
- range(-100, 100, 20) # -100, -80, -60, -40, -20, 0, 20, 40, 60, 80
- range(500, 0, -150)  # 500, 350, 200, 50
- range(-50, 50, -20)  # WORNG
- range(50, -50, 20)   # WRONG

````python
# a, a+d, a+2d, a+3d, ...., a+(n-1)d
# (a + a+(n-1)d) * n/2
# 8, 13, 18, 23, 28, ..., 108
total = 0
for i in range(8, 109, 5):
    total = total + i
print(total)
````

## The while construct/loop
- Conditional loop; the loops based on a condition
- As long as the condition condition is True the loop iterates; once the condition is False the loop ends
- Used to create an infinite loop; for servers, GUI applications
- It is efficient for large datasets

##### Syntax:
```
while bool_exp:
    block
```

- As long as the boolean expression is True the loop iterates executing the block for each iteration.
- Once the bool_exp becomes False, the loop ends

````python
# infinite loop
while 1 < 5:
    print("Hello, world!")
````

````python
# infinite loop
while True:
    print("Hello, Pythonista!")
````

````python
# loops that never run
while 1 > 5:
    print("Good morning!")

# second example    
while False:
    print("Good afternoon!")
````

````python
# printing 0 to 4 using for loop; start, stop, step value
for i in range(5):
    print(i, end=" ")
````

````python
# printing 0 to 4 using while loop
i = 0         # start value
while i < 5:  # stop value
    print(i, end=" ")
    i = i + 1 # step value
````

````python
# sum of n natural numbers
total = 0
i = 1
while i < 11:
    total = total + i
    i += 1      # i = i + 1
print(total)
````

## The `break` statement
- It works with both the `for` and `while` loops equally
- The `break` statements ends a loop early; without going through all the iterations
- Always use the `break` statement inside the `if` statement, because you break a loop for a reason

````python
attendees = ['raju', 'praveen', 'sunil', 'uday', 'prasad', 'mahesh', 'krishna', 'narsing']
for e in attendees:
    print(e)
    if e == 'prasad':
        break
````

````python
values = [100, 88, 43, 49, 59, 99, 101, 300, -15]
for i in values:
    if i % 7 == 0:
        print(i, "divisible by 7.")
        break
````

### The `continue` statement
- To skip an iteration we use `continue` statement
- We write the `continue` statement inside the `if` to skip an iteration for a reason
- When the `continue` statement is executed the control jumps to the beginning to the loop for next iteration

````python
weekdays = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
# skip printing days starting with letter 's' or 't', print remaining
for day in weekdays:
    if day[0] == 's' or day[0] == 't':
        continue
    print(day)
````

````python
quote = "The quick brown fox jumps over the lazy dog."
for l in quote:
    if l != 'a' and l != 'e' and l != 'i' and l != 'o' and l != 'u':
        # print(l, end=' ')     # for printing consonants
        continue
    print(l, end=' ')           # for printing vowels
````
