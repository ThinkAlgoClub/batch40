# User-Defined Functions (UDFs)

#### Types of functions
- Built-in functions: print(), input(), type(), etc.
- Standard Library functions: modules - random, math, decimal, csv, pickle, json
- Third-parth libary functions: numpy, pandas, streamlit, scikit-learn, tensorflow
- User-Defined Functions (UDFs): you create these functions as a user

### UDFs
- They can be created to solve simple to complex problems
- We divide a program into chunks, where each chuck is a function for a small task so that we can reuse the code (reusability)

#### Syntax
```
def func_name([args]):
    block
```

- Use the `def` word to create a function. The `def` stands for definition to define a function
- The `func_name` uses the name rules as variable name
- Inside the parentheses - () - we provide arguments. Argumetns are inputs to a function.
- The arguments [args] - are optional
- The code (block) is indented
- A function doesn't execute the block unless you call (code a calling statement) function
- How do we call a function?
    - **func_name([args])**

````python
# function definition
def welcome():
    print("Hello, Batch-40")
    print("We are learing Functions (UDFs).")

# function call
welcome()
````

````python
# function definition
def welcome(name):
    print("Hello,", name)
    print("How are you doing lately?")
    print("Let's meet again this week.")

welcome("Praveen")
````

````python
# function definition
def welcome(name, age):
    print("Name:", name)
    print("Age:", age)
    print("Let's meet again this week.")

welcome("Satish", 20)
````

````python
# function defintion
def product(name, price, discountPercent):
    print("Name:", name)
    print("Price:", price)
    print("Discount percent:", discountPercent)
    print("Discount amount:", price * discountPercent / 100)
    print("Discount price:", price * ( 1 - discountPercent / 100))

# function call
product("Poko A64", 22000.0, 5)
````

````python
# parameter with a default-value
# makes the parameter optional
# first code required parameters and then code default (optional) parameters
# set a value (default value) for discountPercent parameter in the function defintion first line
# optional parameters can be overwritten by passing a value for it
def product(name, price, discountPercent=2):
    print("Name:", name)
    print("Price:", price)
    print("Discount percent:", discountPercent)
    print("Discount amount:", price * discountPercent / 100)
    print("Discount price:", price * ( 1 - discountPercent / 100))

product('Samung A22', 22000.0)
print()
product("Epson Projector", 42000.0, 10)
print()
product("iPhone", 190000.0, 5)
````

````python
# by defalut parameters are positional
# however, we can change the order of the arguments in the function call using named arguments
def welcome(name, age):
    print("Name:", name)
    print("Age:", age)

# use named arguments in the function call
# order of the arguments doesn't matter
welcome(age=30, name='Pushkal')
````

### The `return` statement
- A function can contain a `return` statement
- A return statement returns a value to the calling statement
- A calling statement is replaced by returned value/object
- And the return ends the function execution

````python
def adding(a, b):
    c = a + b
    print(c)

adding(10, 20)
````

````python
def adding(a, b):
    c = a + b
    return c

adding(10, 20)
````

````python
def adding(a, b):
    c = a + b
    return c

res = adding(10, 20)
print("Result:", res)
````

````python
def sample(x):
    if x == 1:
        print("Good morning!")
    elif x == 2:
        print("Good afternoon!")
    elif x == 3:
        return "Not a valid input"
    print("See you later!")
    print("--------------")

sample(3)
````

## Local and global variables

### Global variables
- Global variables are declared/created ouside of any function
- They are accessbile outside of a function and also inside any function in the program
- By default, inside functions the global variables are read-only
- We can modify a global variable inside a function using the `global` keyword, however, this is not recommended

### Local variables
- A local variable is created/declared inside a function
- It is only accessible inside that particular function
- *Note:* If a local and global variables have same name, then inside it is always local variable that is accessed, outside it is always global variable that is accessed.

````python
def sample():
    print(x)     # global variable

x = 25    # global variable
sample()
````

````python
def sample():
    x = 10     # local 
    print(x)

x = 50       # global
sample()
print(x)
````

````python
def sample(x):  # local
    print(x)    # local

x = 25      # global
sample(x)
print(x)
````

````python
def sample(x):  # local
    x = x * 3
    print(x)    # local

x = 25      # global
sample(x)
print(x)
````

````python
def sample():
    global x      # global - write-mode; not recommended
    x = x * 10
    print(x)

x = 5           # global
sample()
print(x)
````

````python
TAX = 0.18

def tax_amount(amount):
    return amount * TAX

def total_amount(amount):
    return amount + amount * TAX

amount = int(input("Enter invoice amount: "))

print("Amount:", amount)
print("Tax amount:", tax_amount(amount))
print("Total Invoice amount:", total_amount(amount))
````

## Modules

- Module is a Python file
- Filename: temperature.py
- Modulename: temperature
- A Python module is Python file with functions it, without extension
- We use the functions inside a Python by `importing` it
- Syntax for importing
```
import module_name
```
- Syntax for using functions from the module:
```
import module_name
module_name.func_name([args])
```
- Syntax for importing a module with alias (nickname)
```
import module_name as m
m.func_name([args])
```
- Syntax for directly importing functions from a module; not recommended
```
from module_name import func_name1, func_name2
```

````python
import temperature
temperature.to_celsius(100)
````

````python
temperature.to_fahrenheit(38)
````

````python
import temperature as t
t.to_celsius(200)
````

````python
val = random.randrange(1, 10)
guesses = 0

while guesses < 3:
    guess = int(input("Enter a guess: "))
    if guess > val:
        print("Your guess is greater than the value.")
        guesses += 1
    elif guess < val:
        print("Your guess is less than the value.")
        guesses += 1
    else:
        print("You guessed it right!!")
        break

print()
print("Wrong guesses:", guesses)
````
