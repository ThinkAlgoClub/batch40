# Tuples

- Tuple is a collection of sequential (indexed) items.
- Tuples are immutable
- Every item has an index number similar to list object items. Both postive index numbers and negative index numbers.
- **Syntax**:
```
(item1, item2, item3, item4, ...)
```

````python
week_days = ('Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat')
````
