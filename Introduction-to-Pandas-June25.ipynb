## Series

- It is one-dimensional data structure
- It contains two aspects - data and index
- By default, index is a sequence of numbers like list object's index
- However, we can specify the custom index
- Additionally, we can specify the datatype (dtype) of Series

````python
# creating a series of numbers using Series class
pd.Series([1, 2, 3, 4, 5])
````

````python
pd.Series()
````

````python
s1 = pd.Series([10, 20, 30, 40])
s1
````

````python
# np.nan is a missing value; data parameter to specify the data
s2 = pd.Series(data=[100, 200, 300, np.nan, 500])
s2
````

#### np.nan
- Represent a missing value from NumPy
- NaN stands for not a number, but is actually a very small floating-number

#### In a Series or DataFrame we see an Index on the left-side
- Index numbers start from 0 (0 to n-1)
- In Series, index represents individual values
- In DataFrame, index represents a row
- However, a column can also be represent by column_index in DataFrame

````python
# use the index parameter to specify a custom index
s3 = pd.Series(data=['Dosa', 'Upma', 'Idly', 'Puri', 'Vada'],
              index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
s3
````

````python
# A Series from a dictionary
routine = {'Mon':'Python', 'Tue':'NumPy', 'Wed':'pandas', 'Thu':'sk-learn',
          'Fri':'tensorflow'}
# keys become index and values are series values
s4 = pd.Series(routine)
s4
````

````python
# A Series from a set
colors = {'red', 'green', 'blue', 'white', 'black'}
# pd.Series(colors)
pd.Series(list(colors))
````

````python
# A Series from random values generated using NumPy library functions
random = np.random.randint(1, 101, 10)
pd.Series(random)
````

````python
# dtype parameter to specify the data type of a series
pd.Series([1, 2, 3, 4, 5], dtype='int')
````

````python
pd.Series([1, 2, 3, 4, 5], dtype='float')
````
