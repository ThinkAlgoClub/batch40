# WORKING WITH DECIMALS

- We know that floats are approximate numbers, not exact numbers
- So we need exact numbers
- Use `decimal` module, from it use `Decimal` construct to construct a decimal number
- Decimal numbers are exact values

````python
from decimal import Decimal
````

````python
print(Decimal("100.1") + Decimal("100.1") + Decimal("100.1"))
````

### You can mix Decimlas and Integers: legal

````python
100 + Decimal("5.5")
````

### Don't mix Decimals and Floats: illegal

````python
# 100.0 + Decimal("5.5")     # Error
````

### Small program
```
product price
```

````python
from decimal import Decimal
product = 10000
discount_percent = Decimal("0.08")
discounted_price = product * (1 - discount_percent)
tax_percent = Decimal("0.18")
tax_amount = discounted_price * tax_percent
inv_amount = discounted_price + tax_amount
print("Invoice:", inv_amount)
````
