# WORKING WITH SETS

- A set is an unordered collection of items
- A set contains no duplicates
- Set is mutable

#### syntax for creating sets
```
{item1, item2, item3, item4, ...}
```

````python
# a set of colors
colors = {'red', 'green', 'blue', 'yellow', 'brown', 'black', 'white'}
type(colors)
````

````python
a_set = {1, 2, 3, 1, 2, 3, 1, 2, 3}
a_set
````

````python
# Create an empty set using set() function
a_set = set()
a_set
````

````python
colors
````

## SET METHODS

### 1. `add(item)` method
- This method add an specified item to a set

````python
colors
````

````python
colors.add("purple")
colors
````

### 2. `update(items)` method
- We can add more than one item using update method

````python
colors
````

````python
colors.update(['indigo', 'voilet', 'magenta', 'cyan'])
print(colors)
````

### 3. `remove(item)` method
- To romove the specified item from the set

````python
print(colors)
````

````python
colors.remove('yellow')
print(colors)
````

````python
# colors.remove('lightblue')  # KeyError
````

### 4. `discard(item)` method
- Removes an item silently, if the item is not in the set then it doesn't throw an error

````python
print(colors)
````

````python
colors.discard('indigo')
print(colors)
````

````python
colors.discard('yellow')
````

## SET OPERATIONS

### 1. `union(set)` or `|` for union operation

````python
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7}
````

````python
a.union(b)
a | b
````

### 2. `intersection(set)` or '&' for intersection operation

````python
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7}
````

````python
a.intersection(b)
a & b
````

### 3. difference between sets

````python
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7}
````

````python
a - b
````

````python
b - a
````

### 4. `symmetric_differece(set)` operator
- Find the symmetric difference between two sets

````python
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7}
````

````python
a.symmetric_difference(b)
a ^ b
````
