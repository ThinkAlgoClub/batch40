### The 3 read methods
1. read() - It reads entire file as one string
2. readlines() - Reads the file into a list, where each line from the file is an item in the list.
3. readline() - Read the next line; starting from line one

````python
# The read() method - reads entire file as one string
f = open("courses.txt", "r")
print(f.read())
f.close()
````

````python
# readlines() - reads a file into a list
f = open("courses.txt", "r")
print(f.readlines())
f.close()
````

````python
# readline - reads the next line from the file
f = open("courses.txt", "r")
line1 = f.readline()
line2 = f.readline()
line3 = f.readline()
print(line1, line2, line3, sep="")
f.close()
````

````python
# example
years = [2020, 2021, 2022, 2023, 2024]
f = open("years.txt", "w")
for year in years:
    f.write(str(year) + "\n")
f.close()
````

````python
f = open("years.txt", "r")
print(f.read())
f.close()
````

### Use the `for` loop to loop through each line of the file

````python
file_years = []
f = open("years.txt", "r")
for line in f:
    line = line.replace("\n", "")
    line = int(line)
    file_years.append(line)
print(file_years)
f.close()
````

````python
# closing the file automatically
file_years = []
with open("years.txt", "r") as f:
    for line in f:
        line = line.replace("\n", "")
        line = int(line)
        file_years.append(line)
print(file_years)
````

### Use `with` statement to automatically close a file
- Open a file using a with statement then we don't need to use the close() method to close the file.
- When the `with` clause (indented statements) end, the file closes automatically

## WORKING WITH `CSV` FILES

- CSV stands for comma-separated values
- CSV files represent tabular data
- **Sample**
```
empId,name,city,age,dept
1001,rajesh,hyd,35,hr
1002,kumar,amr,39,sales
1003,uday,che,25,marketing
1004,praveen,tri,30,admin
1005,prasad,blr,45,management
```
- Every column is separated by a comma (,)
- Every line ends with a new line (\n) character

- To work with `csv` files use the `csv` module
- To write data to a csv file, we need to create a csv writer object
    - Use writer() function from CSV module, then write use `writerows()` method
- To read data from a csv file, we need to create a reader object
    - Use reader() function from CSV module, then use loop to read each row

````python
import csv

# 5 rows and 4 columns
student_marks = [
    ['name', 'maths', 'physics', 'chemistry'],
    ['praveen', 99, 85, 98],
    ['naveen', 100, 99, 98],
    ['guna', 88, 98, 78],
    ['jaswanth', 91, 92, 93]
]

with open("student_marks.csv", "w", newline="") as f:
    w_obj = csv.writer(f)
    w_obj.writerows(student_marks)
````

````python
import csv

with open("student_marks.csv", "r", newline="") as f:
    r_obj = csv.reader(f)
    for line in r_obj:
        print(line[0] + "\t\t" + line[3])
````

## Binary (non-text) files

- To work with binary files we use `pickle` module
- This moudle is used to work with persistent data
- Use in serilaization and de-serialization of data
- This module has two functions: `dump` to write and `load` to read

````python
import pickle

objs = [
    ['Rajesh', 35, 'hyd', True],
    ['Kiran', 28, 'che', False],
    ['Prasad', 40, 'del', False],
    ['Anand', 45, 'amr', False],
    ['Praveen', 39, 'che', True]
]
````

````python
with open('emps.bin', 'wb') as file:
    pickle.dump(objs, file)
````

````python
# reading from the binary files
with open('emps.bin', 'rb') as file:
    data = pickle.load(file)
    print(data)
````
