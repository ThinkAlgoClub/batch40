# EXCEPTION HANDLING

## What are exceptions?

### Errors vs Exceptions

#### Errors (Syntax)
- When?
    - Before the program runs
- Why?
    - The code is grammatically incorrect for Python
- Can we handle?
    - No - fix the code
- Example:
    - print("hello"
    - (missing parentheses)

#### Exceptions
- When?
    - During execution (runtime)
- Why?
    - The code is valid, but something unexpected happened
- Can we handle?
    - Yes - catch and respond
- Examples:
    - 1/0 (ZeroDivisionError)

**Note**: When an excpetion occurs Python creates an `exception object` and "raises" it. If we don't catch the exception the program terminates with a traceback.

### The `try / except` mechanism

- We use `try` and `except` to tell Python: "Try this block, and if that type of problem occurs, do this other thing instead."

#### Basic Structure
```
try:
    # Code that might go wrong
    risky_operation()
except SomeException:
    # what to do if that specific problem occurs
    recovery_plan()
```

````python
# Catching ZeroDivisionError
try:
    numerator = 10
    denominator = 0
    result = numerator / denominator
    print("Result:", result)
except ZeroDivisionError:
    print("Oops! You can't divide by zero.")

print("Program continues...")
````

````python
# Only the first exception is caught
try:
    x = int("abc")
    y = 1 / 0
except ValueError:
    print("Invalid number format")
````

### Catching Specific Exceptions
- Multiple `except` blocks, exception hierarchy, ordering

````python
# Handling different error types separately
try:
    value = int(input("Enter a number: "))
    result = 100 / value
except ValueError:
    print("That wasn't a valid number!")
except ZeroDivisionError:
    print("Cannot divide by zero!")

print("Carring on...")
````

**Note:** Only the first matching `except` block executes.

````python
# Catching a parent class catches all subclasses
try:
    x = 1 / 0
except ArithmeticError:
    print("Maths error occured")
````

````python
# Always order from most specific to most general
'''
# GOOD
try:
    risky()
except FileNotFoundError:
    print("File missing")
except OSError:          # broader
    print("Some other OS error")

# BAD (broader first makes narrower unreachable)
try:
    risky()
except OSError:
    print("OS Error")
except FileNotFoundError:
    print("File missing")
'''
````

````python
# Web Scraper
import requests
from bs4 import BeautifulSoup

def fetch_title(url):
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parse')
        title = soup.find('title').text
        print("Title:", title)
    except requests.exceptions.Timeout:
        print("Ther server is taking too long. Skipping...")
    except requests.exceptions.HTTPError as e:
        print(f"HTTP Error: {e.response.status_code}")
    except requests.exceptions.ConnectionError:
        print("Could not connect. Check your internet.")
    except AttributeError:
        print("Page has no <title> tag")
````

### The `else` and `finally` clauses
- Success path and guranteed cleanup

#### Full structure
```
try:
    # Risky code
except SpecificError:
    # Recovery
else:
    # Success Celebration
finally:
    # Cleaning - guranteed
 ```

````python
# `finally` guarantees file close

file = None

try:
    file = open("data.txt", "r")
    content = file.read()
    print(len(content), "characters")
except FileNotFoundError:
    print("File does not exist.")
else:
    print("Read successfully.")
finally:
    if file:
        file.close()
        print("File closed.")
````
